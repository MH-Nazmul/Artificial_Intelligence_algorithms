```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/mnt/CSE/CSE Faculty/Level 3 semester 1/Artificial Intelligence/mahbub sir sessional/Churn_Modelling.csv')
```

```python
df.shape
df.head()
```

```python
df.drop(columns = ['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.head()
```

```python
df['Geography'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
df = pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
df.head()
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited'].values

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_trf = scaler.fit_transform(X_train)
X_test_trf = scaler.transform(X_test)
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(11, activation='sigmoid', input_dim=11),
    tf.keras.layers.Dense(11, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,batch_size=50,epochs=100,verbose=1,validation_split=0.2)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred = y_pred.argmax(axis=-1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
